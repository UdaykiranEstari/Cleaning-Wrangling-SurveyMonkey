## Melting the dataset

- <span class="custom-font">the .melt() function is used to transform or reshape a DataFrame from a wide format to a long format, also known as "unpivoting" the data.</span>

- <span class="custom-font"> The wide format refers to a DataFrame where each row represents a unique observation or entity, and each column represents a different variable or feature.</span>

```python
''' 
This function is useful to massage a DataFrame into a format where one or more columns are 
identifier variables (id_vars), while all other columns, considered measured variables (value_vars), 
are “unpivoted” to the row axis, leaving just two non-identifier columns, "variable" and "value". 

'''

identifiers = ['Respondent ID', 'Identify which division you work in.-Response',
       'Identify which division you work in.-Other (please specify)',
       'Which of the following best describes your position level?-Response',
       'Which generation are you apart of?-Response',
       'Please select the gender in which you identify.-Response',
       'Which duration range best aligns with your tenure at your company?-Response',
       'Which of the following best describes your employment type?-Response']

df_melted = df.melt(id_vars=identifiers, var_name='Question-Subquestion',value_name='Answer')

df_melted.head()
```

```python
df2 = pd.read_excel('/Users/arjun/Data Projects/01_SurveyMonkey/001_Input_files/Data - Survey Monkey Output.xlsx',
                   sheet_name='raw_data', header=None, nrows=2)
df2_transposed = df2.T
df2_transposed.rename(columns={0:'Question', 1:'Subquestion'}, inplace=True)
df2_transposed.fillna(method='ffill',inplace=True)
df2_transposed
```

```python

df2_transposed['Question+Subquestion'] = df2_transposed['Question']+'-'+df2_transposed['Subquestion']
df2_transposed
```

```python
df2_final = df2_transposed[df2_transposed['Question'].str.contains('Question')][['Question','Question+Subquestion']]
df2_final
```

```python
# Merging two dataframes and dropping a extra column from the merged dataframe.

dataset_merged = pd.merge(left=df_melted, right=df2_final,how='left',
                          left_on='Question-Subquestion', right_on='Question+Subquestion')
dataset_merged = dataset_merged.drop(columns=['Question+Subquestion'])
dataset_merged.head()
```

```python
# Counting the number of unique respondents for each question and storing the result in a new column "Total Respondants".

responses = dataset_merged[dataset_merged['Answer'].notna()]
respondants = responses.groupby('Question')['Respondent ID'].nunique().reset_index()
respondants.rename(columns={'Respondent ID':'Total Respondants'}, inplace=True)
respondants.head()
```

```python
#The code is merging two datasets (dataset_merged and respondants) on the 'Question' column.

dataset_merged_2 = pd.merge(left=dataset_merged, right=respondants,how='left',
                          left_on='Question', right_on='Question')
dataset_merged_2.head()
```

```python
# Grouping respondents by their answers to the same question-subquestion and counting the number of unique respondents with the same answer.

responses = dataset_merged[dataset_merged['Answer'].notna()]
same_answer = responses.groupby(['Question-Subquestion','Answer'])['Respondent ID'].nunique().reset_index()
same_answer.rename(columns={'Respondent ID':'Same Answer'}, inplace=True)
same_answer
```

```python
# Merging two datasets on the columns 'Question-Subquestion' and 'Answer', filling any missing values with 0.

dataset_merged_3 = pd.merge(left=dataset_merged_2, right=same_answer,how='left',
                          left_on=['Question-Subquestion','Answer'], right_on=['Question-Subquestion','Answer'])
dataset_merged_3['Same Answer'].fillna(0, inplace=True)
dataset_merged_3.head()
```

```python
output_df = dataset_merged_3.copy()
```

```python
# renaming the columns as they sense 

output_df.rename(columns={'Identify which division you work in.-Response' : 'Division',
       'Identify which division you work in.-Other (please specify)' : 'Other division',
       'Which of the following best describes your position level?-Response' : 'Position',
       'Which generation are you apart of?-Response': 'Generation',
       'Please select the gender in which you identify.-Response' : 'Gender',
       'Which duration range best aligns with your tenure at your company?-Response': 'Tenure duration',
       'Which of the following best describes your employment type?-Response':'Employment type'}
       ,inplace=True)
```

```python
# Arranging the data as per final format or user requirement

output_df = output_df[['Respondent ID', 'Division', 'Other division', 'Position', 'Generation',
       'Gender', 'Tenure duration', 'Employment type', 'Question',  'Question-Subquestion',
       'Answer','Total Respondants', 'Same Answer']]
output_df.head()
```

```python
# Respondants vs Division
fig = plt.figure(dpi=200,figsize=(12,7))
output_df[['Division','Total Respondants']].groupby(['Division']).mean().sort_values(
    by='Total Respondants').plot.barh(ax = plt.gca())
```
